```python
import requests

r = requests.get("https://codezine.jp/")

print(type(r))
print(r.status_code)
```

```python
text = r.text
```

```python
text.split("\n")

for line in text.split("\n"):
    if "<title>" in line or "<h" in line:
        print(line.strip())
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(text, "html.parser")

print(soup.title)
print("------------------------------------")
print(soup.h1)
print("------------------------------------")
print(soup.h1.a.img["alt"])
```

```python
atags = soup.find_all("a")

counter = 1

print(len(atags))
    
for tag in atags:

    print("tag.text " + str(counter) + ": " + tag.text)
    print('tag["href"] ' + str(counter) + ": " + tag["href"])
    print("---------------------------------------------------")
    counter += 1
```
